### Import libraries 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

### Importing dataset

```python
df=pd.read_csv('polynomial_regression.csv')
```

```python
df.head()
```

### Splitting dataset

```python
X=df.drop("RBC's (in billions)",axis=1)
Y=df["RBC's (in billions)"]
```

### Training dataset with linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,Y)
```

### Fitting dataset with polynomial features

```python
from sklearn.preprocessing import PolynomialFeatures
reg=PolynomialFeatures(degree=6)
X_poly=reg.fit_transform(X)

lr2=LinearRegression()
lr2.fit(X_poly,Y)
```

### Visualizing Linear Regression results 

```python
plt.scatter(X,Y,color='red')
plt.plot(X,lr.predict(X),color='blue')
plt.xlabel('Features of data scientist')
plt.ylabel("Salary in USD")
plt.title('Linear Regression')
plt.show()
```

### Visualizing predicted results by polynomia regression 

```python
plt.scatter(X,Y,color='red')
plt.plot(X,lr2.predict(X_poly),color='blue')
plt.xlabel("Features of data scientist")
plt.ylabel("Salary in USD")
plt.title('Polynomial Regression')
plt.show()
```

```python
from sklearn.metrics import r2_score
l_r2=r2_score(Y,lr.predict(X))
print("Linear Regression r2 score : ",l_r2)
```

```python
from sklearn.metrics import r2_score
p_r2=r2_score(Y,lr2.predict(X_poly))
print("Polynomial Regression r2 score : ",p_r2)
```
